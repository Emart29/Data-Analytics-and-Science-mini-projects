Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
#import library 
import pandas as pd
print('Done')
```

```python
# Read in the data
schools = pd.read_csv("schools.csv")
print('Done Reading CSV')
```

```python
# To get which NYC schools have the best math result at the 80% level
# First, calculate the percentage as a new column, then filter and sort by it

schools['math_percent'] = (schools['average_math'] / 800) * 100
# Filter schools with math_percent >= 80, then sort by average_math descending

best_math_schools = schools[schools['math_percent'] >= 80][['school_name', 'average_math']].sort_values('average_math', ascending=False)
best_math_schools
```

```python
# Create a new column for the average SAT score across math, reading, and writing
schools['average_score'] = (schools['average_math'] + schools['average_reading'] + schools['average_writing']) / 3

# Create a new column for the total SAT score (sum of math, reading, and writing)
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']

# Select school name and total SAT score, sort by total SAT descending, and get the top 10 schools
top_10_schools = schools[['school_name', 'total_SAT']].sort_values('total_SAT', ascending=False).head(10)
top_10_schools
```

```python
schools.head(5)
```

```python
# Analyze which single borough has the largest standard deviation in the combined SAT score
# Group by borough and calculate required statistics

borough_stats = schools.groupby('borough').agg(
    num_schools=('school_name', 'count'),
    average_SAT=('total_SAT', 'mean'),
    std_SAT=('total_SAT', 'std')
).reset_index()

# Find the borough with the largest std_SAT
max_std_borough = borough_stats.loc[borough_stats['std_SAT'].idxmax()]

# Create the result DataFrame with rounded values
largest_std_dev = pd.DataFrame({
    'borough': [max_std_borough['borough']],
    'num_schools': [round(max_std_borough['num_schools'], 2)],
    'average_SAT': [round(max_std_borough['average_SAT'], 2)],
    'std_SAT': [round(max_std_borough['std_SAT'], 2)]
})

# The borough with the largest standard deviation in total SAT scores is shown below
print('The borough with the largest standard deviation: \n' , largest_std_dev)
```
